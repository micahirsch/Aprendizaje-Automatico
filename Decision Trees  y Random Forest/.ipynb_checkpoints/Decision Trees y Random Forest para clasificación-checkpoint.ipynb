```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
# Modulos específicos para árboles de decisión
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

from sklearn.ensemble import RandomForestClassifier

# Módulos utiles para evaluar el modelo 
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

from sklearn import metrics
```

Para este ejercicio usé el dataset de [Arbolado en espacios verdes](https://data.buenosaires.gob.ar/dataset/arbolado-espacios-verdes), sacado de la página de la Ciudad Autónoma de Buenos Aires. El objetivo es aplicar los métodos Decision Trees y Random Forest a este dataset para ver cuán bien son capaces de clasificar, según la información provista, la especie de árbol.

```python
df = pd.read_csv("arboles.csv")
```

```python
df.head()
```

Información a tener en cuenta sobre las columnas del dataset:
<ul> 
    <li><b>altura_tot</b> se encuentra en metros
    <li><b>diametro</b> es el diámetro del tronco a los 1.3m del piso, y se encuentra expresada en centímetros
    <li><b>inclinacio</b> es la inclinación del árbol dada en grados, siendo 0 un árbol completamente vertical y 90 uno horizontal
</ul>

```python
df.nombre_com.unique()
```

```python
print("Rango de altura de los árboles: [", min(df.altura_tot), ", ", max(df.altura_tot), "]")
print("Rango de diámetro de los árboles: [", min(df.diametro), ", ", max(df.diametro), "]")
print("Rango de inclinación de los árboles: [", min(df.inclinacio), ", ", max(df.inclinacio), "]")
```

```python
colArboles = ['Purple', 'Blue', 'Green', 'Yellow']
nBins = 55

df.pivot(columns = 'nombre_com', values = 'altura_tot').plot.hist(color = colArboles, bins = nBins, alpha = 0.5, edgecolor='black', density = True)
plt.xlabel('Altura (m)')
plt.ylabel('Densidad')
plt.legend()
plt.title('Histograma de altura según la especie')

df.pivot(columns = 'nombre_com', values = 'diametro').plot.hist(color = colArboles, bins = nBins, alpha = 0.5, edgecolor='black', density = True)
plt.xlabel('Diámetro (cm)')
plt.ylabel('Densidad')
plt.legend()
plt.title('Histograma del diámetro según la especie')

df.pivot(columns = 'nombre_com', values = 'inclinacio').plot.hist(color = colArboles, bins = nBins, alpha = 0.5, edgecolor='black', density = True)
plt.xlabel('Inclinación (grados)')
plt.ylabel('Densidad')
plt.legend()
plt.title('Histograma de la inclinación según la especie')
plt.show()
```

Parecería ser que con una sola columna no voy a poder clasificar correctamente el tipo de árbol.

### Decision Trees

```python
# Lista de features que vamos a considerar 
features = ['altura_tot', 'diametro', 'inclinacio']

# Variable a predecir
target = 'nombre_com'

# Construcción de la matriz de features
X = df[features].to_numpy()

# Construcción del vector a predecir
y = df[target].to_numpy()

# Creacion de las matrices de entrenamiento y testeo. 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.10)
```

```python
dtc = DecisionTreeClassifier(max_leaf_nodes = 5)
```

```python
# Entrenamiento del modelo
dtc.fit(X_train, y_train)

# Score del modelo en el conjunto de entrenamiento 
score_in_train = dtc.score(X_train, y_train)

# Score del modelo en el conjunto de testeo
score_in_test = dtc.score(X_test, y_test)

print('Score en el set de entrenamiento: {}'.format(score_in_train))
print('Score en el set de testeo: {}'.format(score_in_test))
```

```python
# Visualizo el árbol de decisión
fig, ax = plt.subplots(figsize = (15, 8))

plot_tree(dtc, feature_names=features, fontsize = 12, filled = True, ax = ax)
```
